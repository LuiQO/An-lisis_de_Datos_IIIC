<h2 style="color : #8B0000"> 1.2 Carga de archivos de Datos </h2>

```python



dfplayers=pd.read_csv("players.csv")
dfplayers = dfplayers.rename(columns={'PLAYER_NAME': 'Nombre del Jugador', 'TEAM_ID': 'TEAM_ID', 'PLAYER_ID': 'ID Jugador', 'SEASON': 'Temporada'})
dfplayers
```

```python
dfplayers.shape
```

```python
dfequipos=pd.read_csv("teams.csv")
dfequipos
```

```python
dfequipos.shape
```

<h2 style="color : #8B0000"> 1.3 Union de Archivos de Datos </h2>

```python
df_left = pd.merge(dfplayers, dfequipos, on='TEAM_ID', how='left')

df_left
```

<h3> <font color='5F9EA0'>2.2 Contar los valores faltantes o perdidos por variable</font> </h3>

```python
valoresperdidos=df_left.isnull()
valoresperdidos.head(15)
```

<h3> <font color='0000FF'>2.3 Análisis de imputación de valores con la media</font> </h3>

```python
avg_norm=df_left['ARENACAPACITY'].astype('float').mean(axis=0)
print("La media de ARENACAPACITY es: ",avg_norm)
```

```python
df_left['ARENACAPACITY'].replace(np.nan,avg_norm,inplace=True)
df_left.head(15)
```

```python
#Contar los valores nulos
valoresperdidos=df_left.isnull()


#Recorre el dataframe con resultado False: para los datos 'No Nulos' y True: para los datos 'Nulos'.
for column in valoresperdidos.columns.values.tolist():
    print(column)
    print(valoresperdidos[column].value_counts())
    print("")
```
